```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv(r"Data\NBA_League_Averages.csv")
```

```python
print(data)
```

```python
df = pd.DataFrame(data)
```

```python
seasons = df['Season']
three_pa = df['3PA']
fga = df['FGA']
```

```python
plt.figure(figsize=(30, 6))
sns.lineplot(x= seasons , y=three_pa, marker='o', color='purple')

plt.title('NBA League Average 3-Point Attempts per Game (2015–2025)', fontsize=14)
plt.xlabel('Seasons', fontsize=12)
plt.ylabel('3-Point Attempts per Game', fontsize=12)
plt.grid(True)
plt.gca().invert_xaxis()
```

```python
#calculates 2point attempts since it is not a direct stat calculated
df = pd.DataFrame(data)
fga = df['FGA']
def calculate_2pa(fga, three_pa):
    """
    Calculate 2-point field goal attempts (2PA).
    
    Parameters:
    fga (int): Total field goal attempts
    three_pa (int): 3-point field goal attempts
    
    Returns:
    int: 2-point field goal attempts
    """
    two_pa = fga - three_pa
    return two_pa
```

```python
print(two_pa)
```

```python
#Visualizes the amount of 2 point attempts taken in per game in the last 20 years
plt.figure(figsize=(30, 6))
sns.lineplot(x= seasons , y=two_pa, marker='o', color='red')

plt.title('NBA League Average 2-Point Attempts per Game (2015–2025)', fontsize=14)
plt.xlabel('Seasons', fontsize=12)
plt.ylabel('2-Point Attempts per Game', fontsize=12)
plt.grid(True)
plt.gca().invert_xaxis()
```

How has the 3-point shot impacted the league over time ?

The first line plot depicits the the amount of 3 point attempts per game in the last 20 years, 
while the next graph depicts the amount of 2 point attempts per game. The 3 point attempts per game 
data is left skewed depicting an increase in 3 point shots, while the 2 point attempts are right 
skewed indicating decrease in 2 point attempts taken. From this data you can conclude that the NBA has 
changed in play styles to which todays game promotes more 3point shooting from players. For future projects I would cross
